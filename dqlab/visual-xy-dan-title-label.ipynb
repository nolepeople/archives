```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("dataset/pgh.csv")
x_label = dataset["NAMA KELURAHAN"]
y_bar = dataset["LAKI-LAKI WNI"]
plt.bar(x=np.arange(len(x_label)),height=y_bar)
plt.xticks(np.arange(len(x_label)),x_label,rotation=90)
plt.ylabel("jumlah penduduk laki laki")
plt.xlabel("nama kelurahan dijkt-p")
plt.title("Jumlah Penduduk Laki- Laki dijkt")
plt.show()
```
